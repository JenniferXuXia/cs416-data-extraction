```python
import pandas as pd
import numpy as np
import plotly as px
import matplotlib as plt
```

```python
df = pd.read_csv('billie_eilish_discography.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# plot the average danceability of each album with stderr bars using plotly express
df.groupby('album').danceability.mean().plot(kind='bar', yerr=df.groupby('album').danceability.std())
```

```python
# plot the average energy of each album with stderr bars using plotly express
df.groupby('album').energy.mean().plot(kind='bar', yerr=df.groupby('album').energy.std())
```

```python
# plot the average valence of each album with stderr bars using plotly express
df.groupby('album').valence.mean().plot(kind='bar', yerr=df.groupby('album').valence.std())
```

```python
# bar chart of valences for each track in HIT ME HARD AND SOFT
df[df['album'] == 'HIT ME HARD AND SOFT'].valence.plot(kind='bar')
```
